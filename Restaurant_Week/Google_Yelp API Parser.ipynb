```python
import pandas as pd
import json
import urllib
import requests

# Import output file from the R script
restaurants = pd.read_csv("C://../restaurants.csv")
restaurants
```

```python
# Removes special characters and places into a new list

queries = []
for i in restaurants["restaurant"]:
    i = str.replace(i,'\xe9','e')
    i = str.replace(i,'\x96', '-')
    i = str.replace(i,'\x92',"'")
    queries.append(i)
queries
```

```python
# Get key from https://console.developers.google.com/apis/
# Using the Google Places API Web Service

api_key = INSERT KEY HERE
service_url = 'https://maps.googleapis.com/maps/api/place/textsearch/json'

ratings = []
address = []
prices = []

# Iterates through restaurants and parses out the Google Rating, business address, and price level (1-4)
for i in queries:
    params = {
        'query': i + ' nyc',
        'indent': True,
        'key': api_key
}
    url = service_url + '?' + urllib.urlencode(params)
    resp = requests.get(url=url, params=params)
    data = json.loads(resp.text)
    try:
        ratings.append(data["results"][0]["rating"])
    except KeyError:
        ratings.append(None)
    try:
        address.append(data["results"][0]["formatted_address"])
    except KeyError:
        address.append(None)
    try:
        prices.append(data["results"][0]["price_level"])
    except KeyError:
        prices.append(None)
```

```python
restaurants['ratings'] = ratings
restaurants['prices'] = prices
restaurants['address'] = address
restaurants
```

```python
# Now we use the Yelp API. Find the documentation here: https://www.yelp.com/developers/documentation/v2/search_api

from yelp.client import Client
from yelp.oauth1_authenticator import Oauth1Authenticator

auth = Oauth1Authenticator(
    consumer_key= INSERT CONSUMER KEY,
    consumer_secret= INSERT CONUMSER SECRET,
    token= INSERT TOKEN,
    token_secret= INSERT TOKEN SECRET
)

yelp_stars = []
latitude = []
longitude = []

# The API will search each restaurant, take the first result, and parse out the star rating, latitude, and longitude
for i in queries:
    params = {
        'term': i,
        'limit': 1
    }

    response = Client(auth).search('New York City', **params)
    try:
        yelp_stars.append(response.businesses[0].rating)
    except IndexError:
        yelp_stars.append(4)
    try:
        latitude.append(response.businesses[0].location.coordinate.latitude)
    except IndexError:
        latitude.append('No latitude available')
    try:
        longitude.append(response.businesses[0].location.coordinate.longitude)
    except IndexError:
        longitude.append('No longitude available')
```

```python
restaurants['yelp_rating'] = yelp_stars
restaurants['description'] = snippet
restaurants['latitude'] = latitude
restaurants['longitude'] = longitude
```

```python
restaurants.to_csv('C://../restaurants_final.csv', encoding = 'utf-8')
```
